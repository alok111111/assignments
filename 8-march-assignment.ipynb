```python
'''Q1. What is the Probability density function?

 The probability  density function defines the probability function representing the
 density of a continuous random variable lying between a specific range of values.in other
 word ,the probability density function produce the likliehood of values of continuous random variable.
 sometimes it is also called a probabiity distribution function.  
'''

'''Q2. What are the types of Probability distribution?
Ans:
   (a.)Normal or gaussion distribution.
   (b.)poission distribution.
   (c.)uniform distribution.
   (d.)bernouli distribution.
   (e.)binomial distribution.
   (f.)log normal distribution.
'''

'''
Q3. Write a Python function to calculate the probability density function of a normal distribution with 
given mean and standard deviation at a given point.



import math

def normal_pdf(mean, std_dev, x):
    exponent = -((x - mean)**2) / (2 * std_dev**2)
    pdf = (1 / (std_dev * math.sqrt(2 * math.pi))) * math.exp(exponent)
    return pdf
mean = 0
std_dev = 1
point = 2

pdf = normal_pdf(mean, std_dev, point)
print(f"PDF at point {point}: {pdf}")





'''
'''Q4. What are the properties of Binomial distribution? Give two examples of events where binomial 
distribution can be applied


Ans: - (a.) For Discrete random variable .
       (b.) Every experiment outcomes is binary.
       (c.) These experiment is performed for n trials.
       eg.
         tossing a coin ten times.
'''

'''Q5. Generate a random sample of size 1000 from a binomial distribution with probability of success 0.4 
and plot a histogram of the results using matplotlib.

import numpy as np
import matplotlib.pyplot as plt

# Set the parameters
n = 1000  # Sample size
p = 0.4   # Probability of success

# Generate the random sample
sample = np.random.binomial(n, p, size=1000)

# Plot the histogram
plt.hist(sample, bins=20, color='skyblue', edgecolor='black')

# Add labels and title
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n=1000, p=0.4)')

# Display the plot
plt.show()

'''

'''Q7. How Binomial distribution different from Poisson distribution
Ans:- Binomial distribution describes the distribution of binary data from a finite sample.
thus it gives the probability of getting r events out of n trials.poission distribution of 
binary data from an infinite sample.thus it gives the probability of getting r events in a population.
'''



'''Q6. Write a Python function to calculate the cumulative distribution function of a Poisson distribution 
with given mean at a given point.

from scipy.stats import poisson

def poisson_cdf(mean, k):
    cdf = poisson.cdf(k, mean)
    return cdf
mean = 3.5
point = 2

cdf = poisson_cdf(mean, point)
print(f"CDF at point {point}: {cdf}")

'''


'''Q8. Generate a random sample of size 1000 from a Poisson distribution with mean 5 and calculate the 
sample mean and variance.


solve:

import numpy as np

# Set the parameters
mean = 5
sample_size = 1000

# Generate the random sample
sample = np.random.poisson(mean, size=sample_size)

# Calculate the sample mean and variance
sample_mean = np.mean(sample)
sample_variance = np.var(sample)

# Print the results
print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)

'''
#Q9. How mean and variance are related in Binomial distribution and Poisson distribution?

'''


Ans:-
   
  In both Binomial and Poisson distributions, the mean and variance are related, but the specific relationship differs between the two distributions.

Binomial Distribution:
In a Binomial distribution, with parameters n (number of trials) and p (probability of success on each trial), the mean (μ) and variance (σ^2) are related as follows:
Mean (μ) = n * p
Variance (σ^2) = n * p * (1 - p)

Here, the mean represents the expected number of successes in n trials, and the variance represents the spread or dispersion of the distribution. As the probability of success (p) increases, the mean and variance increase as well. The relationship between the mean and variance is directly influenced by the probability of success (p) and the number of trials (n).

Poisson Distribution:
In a Poisson distribution, with parameter λ (the average rate of events per unit time or space), the mean (μ) and variance (σ^2) are related as follows:
Mean (μ) = λ
Variance (σ^2) = λ

In this case, the mean and variance are equal, and they both represent the average rate of events occurring in the given time or space. The Poisson distribution is often used to model rare events when the average rate of occurrence is known. When λ is large, the Poisson distribution approaches a normal distribution, where the mean and variance may not be exactly equal, but they are close.

In summary, for the Binomial distribution, the variance depends on both the number of trials and the probability of success, while for the Poisson distribution, the variance is solely determined by the average rate (λ) of events.

'''

'''Q10. In normal distribution with respect to mean position, where does the least frequent data appear?


Ans:-
    In a normal distribution, the least frequent data points appear in the tails of the distribution, which are the regions farther away from the mean. Specifically, the data points that are located in the extreme ends of the distribution, both on the left (negative values) and right (positive values) sides, are less frequent compared to the data points near the mean.

The normal distribution is symmetric, with the mean as its center point. As you move away from the mean in either direction, the density of data points gradually decreases. The distribution follows the famous bell-shaped curve, with the highest density of data points located around the mean. The frequency of data points progressively decreases as you move away from the mean towards the tails of the distribution.

The probability density function (PDF) of the normal distribution approaches zero as you move further away from the mean in both directions. Therefore, the data points located in the tails of the distribution are the least frequent ones.


It is important to note that the exact position of the least frequent data point in a normal distribution depends on the specific parameters of the distribution, such as the mean and standard deviation. The farther the data point is from the mean, the less frequent it will be in a normal distribution.

'''

```
